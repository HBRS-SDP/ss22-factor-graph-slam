## Bayes for distribution

```python
x =  np.random.normal(0, 4, 100000) # random variable with normal distribution and mean =0, std = 4
theta = 4 # considering only one parameter (standard deviation) and mean is fixed
nx, cx ,_ = plt.hist(x, bins = 100) # plotting distribution of RV with 100 bins
```

```python
pdf_x = nx/np.sum(nx) # calculating pdf(x) which is f(x|theta)
cx =np.delete(cx, -1)
plt.plot(cx, pdf_x)
```

```python
theta_dist = np.random.normal(6, 1, 100000)
ntheta, ctheta ,_ = plt.hist(theta_dist, bins=100)
```

```python
pdf_theta = ntheta/ np.sum(ntheta)
ctheta = np.delete(ctheta, -1) 
plt.plot(ctheta, pdf_theta)
```

```python
likelihood = pdf_x
prior = pdf_theta

normalizing_factor =0
for i, j in zip(likelihood, prior):
    normalizing_factor+=(i*j)

plt.plot(cx,likelihood,label="likelihood")
plt.plot(cx, prior, label = "Prior")
plt.legend()
plt.grid()
plt.show()
print("Normalizing factor:", normalizing_factor)
```

```python
posterior = (likelihood*prior) / normalizing_factor
```

```python
plt.plot(cx,likelihood,label="likelihood")
plt.plot(cx, prior, label = "Prior")
plt.plot(cx, posterior, label ="posterior")
plt.legend()
plt.grid()
plt.show()

np.std(ctheta)
```
